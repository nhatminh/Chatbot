# RASA
- RASA has many dependency issues, so it is better to create a virtual envirnoment or run it on google colab

## Rasa files 
- config.yml 
    * all the machine learning settings
- domain.yml
    * overview of intents + responses custom actions
- stories.md
    * examples of intent/action sequences
- nlu.md
    * examples of intents/entities
- You will have to edit these files depending on what time of chatbot you want
- https://rasa.com/docs/rasa/playground check out the document to learn more

## Config.yml
- you can set the types of model you want to build, below is my 
- pipeline:
  - name: WhitespaceTokenizer
  - name: RegexFeaturizer
  - name: LexicalSyntacticFeaturizer
  - name: CountVectorsFeaturizer
  - name: CountVectorsFeaturizer
    analyzer: "char_wb"
    min_ngram: 1
    max_ngram: 4
  - name: DIETClassifier
    epochs: 100
  - name: EntitySynonymMapper
  - name: ResponseSelector
    epochs: 100

```python
import os
import rasa
import nest_asyncio
## this apply the nest_asyncio for rasa to run in the jupyter notebook
nest_asyncio.apply()
```

```python
from rasa.cli.scaffold import create_initial_project
```

```python
project = 'test_project'
create_initial_project(project)
```

```python
os.chdir(project)
print(os.listdir("."))
```

```python
config = "config.yml"
training_files = "data/"
domain = 'domain.yml'
output = "models/"

print(config, training_files, domain, output)
```

```python
model_path = rasa.train(domain, config, [training_files], output)
print(model_path)
```

```python
from rasa.jupyter import chat

endpoints = "endpoints.yml"

chat(model_path, endpoints)
```

```python
## evaluating the model performance
import rasa.data as data

stories_directory, nlu_data_directory = data.get_core_nlu_directories(training_files)

print(stories_directory, nlu_data_directory)
```
